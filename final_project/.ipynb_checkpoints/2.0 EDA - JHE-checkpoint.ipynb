**Why do we mention that this would be a maximization problem?**

The target variable is a quantity and both quantity and price are part of our training data.  This means that the feature revenue must be included.

\begin{equation*}
\ Revenue = Price * Quantity  
\Longrightarrow
\ Quantity =
\frac{Revenue}{Price}
\end{equation*}

### Fact table analysis: Sales Train

Sales train is the data set from where we will extracting more features from.  These are the questions I would like to answer at this EDA estage:

1. **Sales train count questions:**
    1. __Are there different levels of frequencies on dates in sales train data?__
        Yes.  There is a higher number of transactions on January and a lower number in July. Also, there is a noticeable
        downward trend from year 2013 to 2015.  This may signal both increases in quantities or prices thru time
        
    1. __Are there different levels of frequencies on date_block_num in sales data?__
        Yes.  date_block_num corresponds to the months passed from date 1 to our last date in this dataframe.  The 
        pattern described above is easy to see on this visualization.  Therefore, price, quantities and revenues would
        give additional insights to this problem
        
    1. __Are there different levels of frequencies on  shop_id in sales data?__ 
        Yes.  There are some shops that recorded the highest number of transactions in the sales train dataframe.  Shop ids
        26, 38 and 31 have the highest number of transactions
        
    1. __Are there different levels of frequencies on  item_id in sales data?__ 
        Yes.  Most transactions have an id number approximately below 7500.  Also, there is a clear outlier above 20000.
        This may be an error that needs a futher look
        
    1. __Are there different levels of frequencies on  price in sales data?__ 
        No.  Data seems to be normally distributed except for a fringe outlier.  This outlier will be deleted
        
    1. __Are there different levels of frequencies on  revenue in sales data?__ 
        Yes.  Month to month, there are clusters that point out to seasonality.  Highest levels of revenues were observed
        between November 2013 to April 2014.

```python
for i in sales_train.columns:
    sales_train.groupby(i).agg(date= (i, 'count')).plot(figsize= (14,8), title= i, legend= False);
    
```

```python
# Let's order this by date to spot some additional trends
sales_train_ordered = sales_train.sort_values('date').copy()

fig, ax = plt.subplots(figsize= (14,8))
plt.plot(sales_train_ordered['date'], sales_train_ordered['revenue']);
```

I created some additional graphs to visualize the relationship between price and quantity.  I found out that there is an inverse relationship between average price and total quantities, which seems to explain increases in terms of average revenues, as well as decreasing revenues towards the end of this series.  Besides, the Total Revenues per Month graph not only validated that November is the month with the highest revenues, but also a seasonal trend from September to December.  Another seasonal pattern can be seen from January to March.

So, there are some variables that I will incorporate in my analysis apart from Total quantity:
  
  1. Month of year
  1. Year
  1. Average Revenue per month
  
There may other features that I would like to explore:
  1. Difference between Average Revenue and Month Average Revenue
  1. Multiplied and then normalized Month of year and Year
  
    

```python
def plot_analysis_graph(sales_train_ordered, groupedby, metric, func, graph_title):
    """
    This functions is intended to define certain trends in data by month.
    The idea is to visualize certain patterns between quantities, price and revenues
    
    """
    rev_agg = sales_train_ordered.groupby(groupedby).agg(revenue= (metric,func))
    fig,  ax = plt.subplots(figsize = (18,8))
    ax.bar(rev_agg.index, rev_agg['revenue'], align='center')
    ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.grid(which= 'major', axis= 'y', linestyle= '-')
    ax.set_axisbelow(True)
    plt.xlim(-0.5,34)
    ax.text(0.25, 1.12, graph_title,
            transform= ax.transAxes, size= 24, weight= 600, ha= 'left')
    
    fig.savefig('graphs/{}.png'.format(graph_title))
    plt.box(False);
```

```python
# plot the analysis graph created to visualize additional trends on our ordered data set
plot_analysis_graph(sales_train_ordered, 'date_block_num', 'revenue', 'sum', 'Total Revenues per Month')
plot_analysis_graph(sales_train_ordered, 'date_block_num', 'revenue', 'mean', 'Average Revenues per Month')
plot_analysis_graph(sales_train_ordered, 'date_block_num', 'item_cnt_day', 'sum', 'Total Quantities per Month')
plot_analysis_graph(sales_train_ordered, 'date_block_num', 'item_price', 'mean', 'Average Price per Month')
```

```python
items = items[['item_id', 'item_category_id']]
items =pd.merge(items, item_categories_translated, on = 'item_category_id', how = 'left')
```

```python
def plot_cities_and_product_categories(sales_train_ordered, groupedby, metric, func, graph_title):
    """
    This functions is intended to define certain trends in data by month.
    The idea is to visualize certain patterns between quantities, price and revenues
    
    """
    rev_agg = sales_train_ordered.groupby(groupedby).agg(revenue= (metric,func)).sort_values(by = metric)
    fig,  ax = plt.subplots(figsize = (18,8))
    ax.barh(rev_agg.index, rev_agg[metric], align='center')
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.set_axisbelow(True)
    ax.text(0.25, 1.12, graph_title,
            transform= ax.transAxes, size= 24, weight= 600, ha= 'left')
    
    fig.savefig('graphs/{}.png'.format(graph_title))
    plt.box(False);
    

def corr_matrix_competition(df, title):
    """
    This functions is intended to graph the correlations of a given dataframe.
    
    """

    fig, ax = plt.subplots(figsize = (14,8))
    cax = ax.matshow(df.corr())

    fig.colorbar(cax)

    names = df.select_dtypes(exclude = ['datetime', 'timedelta','category']).columns
    ticks = np.arange(len(names))

    plt.xticks(ticks, names, rotation = 'vertical')
    plt.yticks(ticks, names)
    
    ax.text(0, 1.3, title,
        transform= ax.transAxes, size= 24, weight= 600, ha= 'left')

    fig.savefig('graphs/{}.jpg'.format(title), bbox_inches= 'tight', pad_inches= 0);
```

```python
plot_cities_and_product_categories(sales_train_ordered, 'city_en', 'revenue', 'sum', 'Total Revenues per City')
plot_cities_and_product_categories(sales_train_ordered, 'sub_category_1_en', 'revenue', 'sum', 'Total Revenues per Product Category')
corr_matrix_competition(sales_train_ordered, 'Correlations in the Sales Dataset')
```
